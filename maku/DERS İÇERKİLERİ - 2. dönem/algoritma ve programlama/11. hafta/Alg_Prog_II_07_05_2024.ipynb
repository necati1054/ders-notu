# AĞAÇ ALGORİTMALARI

```python
# Basit bir karar ağacı algoritması

def karar_agaci(skor):
    if skor >= 80:
        return 'Futbol'
    elif skor >= 60:
        return 'Basketbol'
    elif skor >= 40:
        return 'Voleybol'
    else:
        return 'Yüzme'

# Oyuncunun skorunu belirleme
oyuncu_skoru = 75

# Karar ağacı algoritmasını kullanarak tahmin yapma
tahmin_edilen_spor = karar_agaci(oyuncu_skoru)

# Sonucu ekrana yazdırma
print(f"Oyuncunun skoru {oyuncu_skoru} olduğu için tercih edilen spor: {tahmin_edilen_spor}")
```

Bu kod örneğinde, karar_agaci adlı bir fonksiyon tanımlanmıştır. Bu fonksiyon, oyuncunun skorunu alır ve belirli bir eşik değere göre hangi sporu tercih edeceğini belirler. Eşik değerlerine göre, futbol için 80, basketbol için 60, voleybol için 40 olarak belirlenmiştir.

Daha sonra, bir oyuncunun skoru belirlenir (oyuncu_skoru) ve bu skor karar_agaci fonksiyonuna geçirilir. Karar ağacı algoritması, verilen skora göre hangi sporun tercih edileceğini belirler.

Son olarak, tahmin edilen spor ekrana yazdırılır. Örneğin, eğer oyuncunun skoru 75 ise, program "Oyuncunun skoru 75 olduğu için tercih edilen spor: Basketbol" çıktısını verecektir.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def karar_agaci(skor):
    if skor >= 80:
        return 'Futbol'
    elif skor >= 60:
        return 'Basketbol'
    elif skor >= 40:
        return 'Voleybol'
    else:
        return 'Yüzme'

def draw_decision_tree():
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.axis('off')

    # Karar düğümleri
    ax.text(0.5, 0.8, "Skor >= 80", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='blue', boxstyle='round,pad=0.5'))
    ax.text(0.2, 0.5, "Futbol", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='green', boxstyle='round,pad=0.5'))
    ax.text(0.5, 0.5, "60 <= Skor < 80", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='blue', boxstyle='round,pad=0.5'))
    ax.text(0.8, 0.5, "Basketbol", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='green', boxstyle='round,pad=0.5'))
    ax.text(0.5, 0.2, "40 <= Skor < 60", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='blue', boxstyle='round,pad=0.5'))
    ax.text(0.5, 0.1, "Voleybol", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='green', boxstyle='round,pad=0.5'))
    ax.text(0.5, 0.0, "Skor < 40", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='blue', boxstyle='round,pad=0.5'))
    ax.text(0.5, -0.1, "Yüzme", ha='center', va='center', fontsize=12, bbox=dict(facecolor='lightgreen', edgecolor='green', boxstyle='round,pad=0.5'))

    # Bağlantılar
    ax.plot([0.5, 0.2], [0.7, 0.5], color='black', linestyle='-', linewidth=2)
    ax.plot([0.5, 0.8], [0.7, 0.5], color='black', linestyle='-', linewidth=2)
    ax.plot([0.5, 0.5], [0.6, 0.2], color='black', linestyle='-', linewidth=2)
    ax.plot([0.5, 0.5], [0.4, 0.1], color='black', linestyle='-', linewidth=2)

    plt.title("Karar Ağacı Görselleştirmesi")
    plt.show()

# Karar ağacı görselleştirmesi
draw_decision_tree()
```

```python
import matplotlib.pyplot as plt

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

# Ağaç yapısını oluşturma
root = Node("A")
root.add_child(Node("B"))
root.add_child(Node("C"))
root.children[0].add_child(Node("D"))
root.children[0].add_child(Node("E"))
root.children[1].add_child(Node("F"))

# Ağacı görselleştirme
def visualize_tree(node, ax, x=0, y=0, step_x=1, step_y=1):
    if not node:
        return

    ax.text(x, y, node.value, fontsize=12, ha='center', va='center', bbox=dict(facecolor='lightblue', edgecolor='blue', boxstyle='round,pad=0.5'))
    children_count = len(node.children)
    if children_count > 0:
        next_step_x = step_x / children_count
        next_x = x - (step_x / 2)
        for child in node.children:
            ax.plot([x, next_x], [y, y - step_y], color='black', linestyle='-', linewidth=2)
            visualize_tree(child, ax, x=next_x, y=y - step_y, step_x=next_step_x, step_y=step_y)
            next_x += next_step_x

fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')

visualize_tree(root, ax)

plt.title("Ağaç Yapısının Görselleştirilmesi")
plt.show()
```
