## Veri Yapıları

Bu bölüm, daha önce öğrendiğiniz bazı şeyleri daha ayrıntılı olarak açıklamakta ve bazı yeni şeyler de eklemektedir.

### Listeler Üzerine
Liste veri türünün bazı yöntemleri daha vardır. Liste nesnelerinin tüm metotları şunlardır:

#### list.append(x)
Listenin sonuna bir öğe ekleyin. a[len(a):] = [x] ile eş değerdir.

#### list.extend(iterable)
Listeyi iterable’daki tüm öğeleri ekleyerek genişletin. a[len(a):] = iterable ile eş değerdir.

#### list.insert(i, x)
Verilen pozisyona bir öğe ekleyin. İlk argüman, daha önce ekleyeceğiniz öğenin indeksidir, bu nedenle a.insert(0, x) listenin önüne ekler ve a.insert(len(a), x) komutu a.append(x) komutu ile eş değerdir.

#### list.remove(x)
Değeri x’e eşit olan ilk öğeyi listeden kaldırır. Böyle bir öğe yoksa ValueError hatası ortaya çıkar.

#### list.pop([i])
Remove the item at the given position in the list, and return it. If no index is specified, a.pop() removes and returns the last item in the list. It raises an IndexError if the list is empty or the index is outside the list range.

#### list.clear()
Listeden tüm öğeleri kaldırır. del a[:] ile eş değerdir.

#### list.index(x[, start[, end]])
Değeri x’e eşit olan ilk öğenin listedeki sıfır tabanlı indeksini döndürür. Böyle bir öğe yoksa ValueError hatası ortaya çıkar.

İsteğe bağlı start ve end argümanları dilim gösteriminde olduğu gibi yorumlanır ve aramayı listenin belirli bir alt dizisiyle sınırlamak için kullanılır. Döndürülen dizin, start bağımsız değişkeni yerine tam dizinin başlangıcına göre hesaplanır.

#### list.count(x)
Listede x öğesinin kaç kez göründüğünü döndürür.

#### list.sort(*, key=None, reverse=False)
Listenin öğelerini yerinde sıralayın (argümanlar sıralama özelleştirmesi için kullanılabilir

#### list.reverse()
Listenin öğelerini yerinde ters çevirir.

#### list.copy()
Listenin yüzeysel bir kopyasını döndürür. a[:] ile eş değerdir.

```python
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']
fruits.count('apple')
```

## Stack (Yığın)

Tanım: Yığın, elemanların son giren ilk çıkan (Last In, First Out - LIFO) prensibine göre düzenlendiği bir veri yapısıdır. Yani en son eklenen eleman, yığından ilk çıkarılan elemandır.

### Operasyonlar:

push(item): Yığına bir eleman ekler. Bu işlem yığının en üstüne eleman ekler.

pop(): Yığından en üstteki elemanı çıkarır ve geri döndürür. Yığın boşsa hata verebilir.

peek(): Yığının en üstündeki elemanı döndürür, ancak yığından çıkarmaz.

isEmpty(): Yığının boş olup olmadığını kontrol eder.

size(): Yığının eleman sayısını döndürür.







### Uygulama Alanları:

Tarayıcı geçmişi (geri butonu işlevi)
Derinlik öncelikli arama (depth-first search) algoritmalarında kullanılabilir.
İşlem sırasında geçici verilerin saklanması.

Fonksiyon çağrılarının geri dönüş değerlerini saklamak için kullanılabilir.

Yığın, elemanların son giren ilk çıkan (Last In, First Out - LIFO) prensibine göre düzenlendiği bir veri yapısıdır. Yığın veri yapısında elemanlar yalnızca yığının en üstüne eklenebilir veya en üstünden çıkarılabilir.

Python'da stack, listeleri kullanarak kolayca uygulanabilir. append() yöntemi ile eleman ekleme ve pop() yöntemi ile eleman çıkarma işlemleri kullanılabilir.

```python
stack = []

# Eleman ekleme
stack.append(1)
stack.append(2)
stack.append(3)
stack
```

```python
# Eleman çıkarma
removed_item = stack.pop()
print("Çıkarılan eleman:", removed_item)
```

## Queue (Kuyruk)

Tanım: Kuyruk, elemanların ilk giren ilk çıkan (First In, First Out - FIFO) prensibine göre düzenlendiği bir veri yapısıdır. Yani, kuyruğa ilk eklenen eleman, kuyruktan ilk çıkarılan elemandır.

### Operasyonlar:

enqueue(item): Kuyruğa bir eleman ekler. Bu işlem kuyruğun sonuna eleman ekler.

dequeue(): Kuyruğun başındaki elemanı çıkarır ve geri döndürür. Kuyruk boşsa hata verebilir.

peek(): Kuyruğun başındaki elemanı döndürür, ancak kuyruktan çıkarmaz.

isEmpty(): Kuyruğun boş olup olmadığını kontrol eder.

size(): Kuyruğun eleman sayısını döndürür.





### Uygulama Alanları:

İşlem sırasında bekleyen işlerin veya görevlerin yönetimi (Örneğin, iş parçacığı havuzları)

Veri iletimi veya işleme sırasında geçici verilerin saklanması

Genellikle genişlik öncelikli arama (breadth-first search) algoritmalarında kullanılır.

Kuyruk, elemanların ilk giren ilk çıkan (First In, First Out - FIFO) prensibine göre düzenlendiği bir veri yapısıdır. Kuyrukta elemanlar sırayla eklenir ve sırayla çıkarılır.

Python'da queue veri yapısını queue modülüyle kullanabilirsiniz.

```python
from queue import Queue

queue = Queue()

# Eleman ekleme
queue.put(1)
queue.put(2)
queue.put(3)

print(queue.queue)
```

```python
from queue import Queue

# Queue objesi oluşturma
queue = Queue()

# Eleman ekleme
queue.put(1)
queue.put(2)
queue.put(3)

print("Front of queue:", queue.queue[0])  # 1
print("Size of queue:", queue.qsize())  # 3

# Eleman çıkarma
while not queue.empty():
    print("Removed item:", queue.get())
```

queue.get() ifadesi, Python'da kuyruk (queue) veri yapısından bir elemanı çıkarmak için kullanılan bir yöntemdir. Bu yöntem kuyruğun başındaki elemanı kuyruktan çıkarır ve bu elemanı döndürür.

## Deque (Double Ended Queue - Çift Uçlu Kuyruk)

Tanım: Deque, hem kuyruk hem de yığın işlevselliğini bir arada sunan ve elemanların hem baş hem de son kısımlarından eklenebileceği veya çıkarılabileceği bir veri yapısıdır. Bu özellik, hem FIFO (First In, First Out) hem de LIFO (Last In, First Out) yöntemlerini destekler.

Ekleme ve silme işlemlerinin başlangıçta olduğu kadar sonda da yapılabileceği özel bir tür kuyruk.

Deque'nin bağlantı listesi temsili, her düğümün önceki düğümün yanı sıra sonraki düğüme de işaret edeceği şekildedir. Böylece ekleme ve silme işlemleri hem başlangıçta hem de sonda sabit zaman alır.




### Operasyonlar:

append(item): Dequenin sağ ucuna bir eleman ekler.

appendleft(item): Dequenin sol ucuna bir eleman ekler.

pop(): Dequenin sağ ucundaki elemanı çıkarır ve geri döndürür.

popleft(): Dequenin sol ucundaki elemanı çıkarır ve geri döndürür.

peek(): Dequenin sağ ucundaki elemanı döndürür, ancak çıkarmaz.

peekleft(): Dequenin sol ucundaki elemanı döndürür, ancak çıkarmaz.

isEmpty(): Dequenin boş olup olmadığını kontrol eder.

size(): Dequenin eleman sayısını döndürür.


Bunları yığın ve kuyruk olarak çalışacak şekilde ayarlamak için deque'in işlevleri aşağıda listelenmiştir.






### Uygulama Alanları:

Çift yönlü veri işleme gerektiren algoritmalar.
Kuyruk ve yığın işlevselliğini aynı anda gerektiren durumlar.
Bellek kullanımı ve performans açısından avantajlı olan senaryolar.

Deque, hem kuyruk hem de yığın işlevselliğini bir arada sunan bir veri yapısıdır. Hem kuyruk gibi elemanları sırayla ekleyip sırayla çıkarabilir, hem de yığın gibi elemanları en üstten veya en alttan ekleyip çıkarabilirsiniz.

Python'da deque veri yapısını collections modülü altında bulabilirsiniz.

```python
from collections import deque

# Deque objesi oluşturma
deque_obj = deque()

# Eleman ekleme
deque_obj.append(1)  # Sağa ekleme
deque_obj.appendleft(2)  # Sola ekleme

print("Right end of deque:", deque_obj[-1])  # 1
print("Left end of deque:", deque_obj[0])    # 2
print("Size of deque:", len(deque_obj))      # 2

# Eleman çıkarma
removed_item = deque_obj.pop()      # Sağdan çıkarma
removed_item_left = deque_obj.popleft()  # Soldan çıkarma

print("Removed from right end:", removed_item)
print("Removed from left end:", removed_item_left)
```

Yukarıdaki örnekte, collections modülünden deque (çift uçlu kuyruk) veri yapısını kullanarak örnek bir deque oluşturulmuş ve çeşitli işlemler gerçekleştirilmiştir. Deque, her iki uçtan eleman eklemeye veya çıkarmaya izin verir, böylece hem stack (yığın) hem de queue (kuyruk) işlevselliğini sağlar. Bu özellik, çeşitli senaryolarda kullanımını kolaylaştırır ve performans avantajları sunar.

```python
from collections import deque

deque_obj = deque()

# Eleman ekleme
deque_obj.append(1)  # Sağa ekleme
deque_obj.appendleft(2)  # Sola ekleme

# Eleman çıkarma
removed_item = deque_obj.pop()  # Sağdan çıkarma
removed_item_left = deque_obj.popleft()  # Soldan çıkarma

print("Sağdan çıkarılan eleman:", removed_item)
print("Soldan çıkarılan eleman:", removed_item_left)
```

Bu örneklerde, Python'daki farklı veri yapılarını kullanarak stack, queue ve deque kavramlarını açıklandı. Her biri farklı amaçlar için kullanılır ve uygun senaryolarda tercih edilirler.
